```python
#month column
df['Month'] = pd.DatetimeIndex(df['START_DATE']).month
```

```python
dic = {1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'june',7:'july',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'}
```

```python
df['Month'] = df['Month'].map(dic)
```

```python
df
```

```python
sns.countplot(df['Month'],order = df['Month'].value_counts().index)
```

```python
#Round trip same start location and same stop location
#which month is having more round trips
for i in df['START']:
    for j in df['STOP']:
        if i==j:
            roundtrip = 'yes'
        else:
            roundtrip = 'no'
```

```python
roundtrip = df['START'] == df['STOP']
```

```python
roundtrip
```

```python
df.insert(6,'ROUNDTRIP',roundtrip)
```

```python
df
```

```python
res = df.groupby(['Month'])['ROUNDTRIP'].sum()
a = dict(res)
res
```

```python
a = df['ROUNDTRIP'].value_counts()
# so there are 288 roundtrips and it is 24.9% of total trips 
a
```

```python
plt.pie(a,labels = dict(a).keys(),autopct = '%.1f%%',explode = [0,0.1])
plt.title('ROUND TRIP')
plt.show()
```

```python
name = list(a.keys())
value = list(a.values())
plt.bar(range(len(a)), value, tick_label=name)
# there are 288 roundtrips and it is 24.9% of total trips
# There are more roundtrips in december
# The least roundtrips are in april
# In may there are no RoundTirps
```

```python
plt.barh(name, value)
for i, v in enumerate(value):
    plt.text(v,i,str(v))
plt.show()
    
```
